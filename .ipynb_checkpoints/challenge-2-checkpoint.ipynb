# Challenge 4 - Handle `WHOIS_*` Categorical Data

There are several categorical columns we need to handle. These columns are:

* `URL`
* `CHARSET`
* `SERVER`
* `WHOIS_COUNTRY`
* `WHOIS_STATEPRO`
* `WHOIS_REGDATE`
* `WHOIS_UPDATED_DATE`

How to handle string columns is always case by case. Let's start by working on `WHOIS_COUNTRY`. Your steps are:

1. List out the unique values of `WHOIS_COUNTRY`.
1. Consolidate the country values with consistent country codes. For example, the following values refer to the same country and should use consistent country code:
    * `CY` and `Cyprus`
    * `US` and `us`
    * `SE` and `se`
    * `GB`, `United Kingdom`, and `[u'GB'; u'UK']`

#### In the cells below, fix the country values as intructed above.

```python
# Show all unique values in WHOIS_COUNTRY to see what's messy
websites['WHOIS_COUNTRY'].unique()
```

```python
# Convert all to uppercase for consistency
websites['WHOIS_COUNTRY'] = websites['WHOIS_COUNTRY'].str.upper()

# Replace variants with standardized country codes
websites['WHOIS_COUNTRY'] = websites['WHOIS_COUNTRY'].replace({
    "[U'GB'; U'UK']": 'UK',
    'UNITED KINGDOM': 'UK',
    'GB': 'UK',
    'U.K.': 'UK',
    
    'UNITED STATES': 'US',
    'U.S.': 'US',
    'USA': 'US',
    
    'CYPRUS': 'CY'
})

# Check unique values again
websites['WHOIS_COUNTRY'].unique()
```

Since we have fixed the country values, can we convert this column to ordinal now?

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
# Show a bar chart of the top 20 WHOIS_COUNTRY values
websites['WHOIS_COUNTRY'].value_counts().head(20).plot(kind='bar', figsize=(12, 6), title='Top WHOIS_COUNTRY values')
```

#### After verifying, now let's keep the top 10 values of the column and re-label other columns with `OTHER`.

```python
# Get the 10 most frequent countries
top_10_countries = websites['WHOIS_COUNTRY'].value_counts().head(10).index

# Replace all other countries with 'OTHER'
websites['WHOIS_COUNTRY'] = websites['WHOIS_COUNTRY'].apply(lambda x: x if x in top_10_countries else 'OTHER')
```

Now since `WHOIS_COUNTRY` has been re-labelled, we don't need `WHOIS_STATEPRO` any more because the values of the states or provinces may not be relevant any more. We'll drop this column.

In addition, we will also drop `WHOIS_REGDATE` and `WHOIS_UPDATED_DATE`. These are the registration and update dates of the website domains. Not of our concerns.

#### In the next cell, drop `['WHOIS_STATEPRO', 'WHOIS_REGDATE', 'WHOIS_UPDATED_DATE']`.

```python
# Drop columns that are no longer useful for the analysis
websites = websites.drop(['WHOIS_STATEPRO', 'WHOIS_REGDATE', 'WHOIS_UPDATED_DATE'], axis=1)
```

# Challenge 5 - Handle Remaining Categorical Data & Convert to Ordinal

Now print the `dtypes` of the data again. Besides `WHOIS_COUNTRY` which we already fixed, there should be 3 categorical columns left: `URL`, `CHARSET`, and `SERVER`.

```python
# Show data types to confirm which columns are still categorical
websites.dtypes
```

#### `URL` is easy. We'll simply drop it because it has too many unique values that there's no way for us to consolidate.

```python
# Drop the URL column because it has too many unique values
websites = websites.drop('URL', axis=1)
```

#### Print the unique value counts of `CHARSET`. You see there are only a few unique values. So we can keep it as it is.

```python
# Show the unique values in CHARSET
websites['CHARSET'].value_counts()
```

```python
# Normalize CHARSET values (lowercase)
websites['CHARSET'] = websites['CHARSET'].str.lower()

# Group similar or uncommon values
websites['CHARSET'] = websites['CHARSET'].replace({
    'utf8': 'utf-8',
    'iso8859-1': 'iso-8859-1',
    'iso-8859': 'iso-8859-1',
    'windows-1251': 'other',
    'windows-1252': 'other',
    'unknown': 'unknown',
    'us-ascii': 'us-ascii'
})
```

```python
# Check cleaned CHARSET values
websites['CHARSET'].value_counts()
```

`SERVER` is a little more complicated. Print its unique values and think about how you can consolidate those values.

#### Before you think of your own solution, don't read the instructions that come next.

```python
# Show all unique SERVER values (before simplification)
websites['SERVER'].value_counts()
```

![Think Hard](../images/think-hard.jpg)

```python
# After inspecting the SERVER column, we notice that although it has 239 unique values, many of them are simply different versions or formats of a few major server types.
# To simplify the data, we will group all servers that contain the words: 'apache', 'microsoft', or 'nginx' (case-insensitive) into their respective categories.
# Everything else will be labeled as 'Other'.
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
# Convert all SERVER values to lowercase to ensure consistent string matching
websites['SERVER'] = websites['SERVER'].str.lower()

# Define a function to categorize server types
def classify_server(value):
    if 'microsoft' in value:
        return 'Microsoft'
    elif 'apache' in value:
        return 'Apache'
    elif 'nginx' in value:
        return 'nginx'
    else:
        return 'Other'

# Apply the function to the SERVER column
websites['SERVER'] = websites['SERVER'].apply(classify_server)
```

```python
# Show the final value counts after simplification
websites['SERVER'].value_counts()
```

OK, all our categorical data are fixed now. **Let's convert them to ordinal data using Pandas' `get_dummies` function ([documentation](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)). Also, assign the data with dummy values to a new variable `website_dummy`.**

```python

# pd.get_dummies() creates a new binary (0/1) column for each category. For example, if SERVER has values ['Apache', 'nginx', 'Microsoft', 'Other'], it will create new columns like SERVER_Apache, SERVER_nginx, etc.

# We assign the transformed dataset to a new variable called website_dummy
website_dummy = pd.get_dummies(websites)

```

Now, inspect `website_dummy` to make sure the data and types are intended - there shouldn't be any categorical columns at this point.

# Challenge 6 - Modeling, Prediction, and Evaluation

We'll start off this section by splitting the data to train and test. **Name your 4 variables `X_train`, `X_test`, `y_train`, and `y_test`. Select 80% of the data for training and 20% for testing.**

```python
from sklearn.model_selection import train_test_split

# First, we separate the features (X) and the target variable (y)
X = website_dummy.drop('Type', axis=1)  # All columns except 'Type'
y = website_dummy['Type']              # The target column (0 = benign, 1 = malicious)

# Scale the features before splitting
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Now we split the data into training and testing sets (80% train, 20% test)
# We also use random_state=0 to make the results reproducible
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)
```

#### In this lab, we will try two different models and compare our results.

The first model we will use in this lab is logistic regression. We have previously learned about logistic regression as a classification algorithm. In the cell below, load `LogisticRegression` from scikit-learn and initialize the model.

```python
# Import the LogisticRegression model from scikit-learn
from sklearn.linear_model import LogisticRegression

# Initialize the logistic regression model
log_model = LogisticRegression(max_iter=1000, random_state=0)

```

Next, fit the model to our training data. We have already separated our data into 4 parts. Use those in your model.

```python
# Fit (train) the model on the training data
log_model.fit(X_train, y_train)

```

finally, import `confusion_matrix` and `accuracy_score` from `sklearn.metrics` and fit our testing data. Assign the fitted data to `y_pred` and print the confusion matrix as well as the accuracy score

```python
# Import metrics to evaluate the model
from sklearn.metrics import confusion_matrix, accuracy_score

# Use the model to predict the labels for the test set
y_pred = log_model.predict(X_test)

# Print the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Print the accuracy score
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

```

What are your thoughts on the performance of the model? Write your conclusions below.

```python
# The logistic regression model achieved an accuracy of approximately 92%.
# From the confusion matrix, we see it correctly predicted:
# - 311 benign websites (True Negatives)
# - 18 malicious websites (True Positives)
# However, it misclassified:
# - 5 benign websites as malicious (False Positives)
# - 23 malicious websites as benign (False Negatives)
# This means that although the overall performance is good, the model is more likely to miss some malicious websites than to falsely flag a benign one.

```

#### Our second algorithm is is K-Nearest Neighbors. 

Though is it not required, we will fit a model using the training data and then test the performance of the model using the testing data. Start by loading `KNeighborsClassifier` from scikit-learn and then initializing and fitting the model. We'll start off with a model where k=3.

```python
# Import the KNN model
from sklearn.neighbors import KNeighborsClassifier

# Initialize the model with k=3
knn_model = KNeighborsClassifier(n_neighbors=3)

# Fit (train) the model
knn_model.fit(X_train, y_train)

```

To test your model, compute the predicted values for the testing sample and print the confusion matrix as well as the accuracy score.

```python
# Import metrics
from sklearn.metrics import confusion_matrix, accuracy_score

# Predict test data
y_pred_knn = knn_model.predict(X_test)

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))

# Print accuracy
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred_knn))

```

```python
# The KNN model with k=3 achieved an accuracy of approximately 93.56%.
# According to the confusion matrix:
# - 302 benign websites were correctly classified (true negatives).
# - 32 malicious websites were correctly identified (true positives).
# - 14 benign websites were incorrectly flagged as malicious (false positives).
# - 9 malicious websites were missed (false negatives).

# Compared to the logistic regression model (accuracy â‰ˆ 92.16%), KNN performed slightly better.
```

# Bonus Challenge - Feature Scaling

Problem-solving in machine learning is iterative. You can improve your model prediction with various techniques (there is a sweetspot for the time you spend and the improvement you receive though). Now you've completed only one iteration of ML analysis. There are more iterations you can conduct to make improvements. In order to be able to do that, you will need deeper knowledge in statistics and master more data analysis techniques. In this bootcamp, we don't have time to achieve that advanced goal. But you will make constant efforts after the bootcamp to eventually get there.

However, now we do want you to learn one of the advanced techniques which is called *feature scaling*. The idea of feature scaling is to standardize/normalize the range of independent variables or features of the data. This can make the outliers more apparent so that you can remove them. This step needs to happen during Challenge 6 after you split the training and test data because you don't want to split the data again which makes it impossible to compare your results with and without feature scaling. For general concepts about feature scaling, click [here](https://en.wikipedia.org/wiki/Feature_scaling). To read deeper, click [here](https://medium.com/greyatom/why-how-and-when-to-scale-your-features-4b30ab09db5e).

In the next cell, attempt to improve your model prediction accuracy by means of feature scaling. A library you can utilize is `sklearn.preprocessing.RobustScaler` ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html)). You'll use the `RobustScaler` to fit and transform your `X_train`, then transform `X_test`. You will use logistic regression to fit and predict your transformed data and obtain the accuracy score in the same way. Compare the accuracy score with your normalized data with the previous accuracy data. Is there an improvement?

```python
# Import RobustScaler
from sklearn.preprocessing import RobustScaler

# Initialize the scaler
robust_scaler = RobustScaler()

# Fit and transform X_train
X_train_scaled = robust_scaler.fit_transform(X_train)

# Transform X_test (important: never fit on test data!)
X_test_scaled = robust_scaler.transform(X_test)
```

```python
# Import and initialize logistic regression
from sklearn.linear_model import LogisticRegression

log_model_robust = LogisticRegression(max_iter=1000, random_state=0)

# Fit the model on the scaled training data
log_model_robust.fit(X_train_scaled, y_train)
```

```python
# Import evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score

# Predict on the scaled test set
y_pred_robust = log_model_robust.predict(X_test_scaled)

# Show confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_robust))

# Show accuracy score
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred_robust))
```

```python
# After applying both StandardScaler and RobustScaler, the final accuracy of the logistic regression model remained almost the same (~92.15%).
# With StandardScaler, the model predicted slightly fewer false positives (5 instead of 7), but also slightly more false negatives (23 instead of 21).
# RobustScaler improved the number of correctly predicted malicious websites (true positives: 20 vs 18), which may be important in security-related contexts.
# Overall, both scalers led to very similar performance in this case.
```
